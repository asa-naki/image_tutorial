# 01 — 画像の入出力・表示

## 目標
- `cv2.imread` で画像を読み込む (BGR / グレースケール / 未変換)
- `cv2.imwrite` でファイルに保存する
- matplotlib でインライン表示する

## ポイント
- OpenCV の色順は **BGR** (≠ RGB)。matplotlib に渡す前に `[:, :, ::-1]` または `cv2.cvtColor(..., COLOR_BGR2RGB)` で反転する。
- `imread` は失敗しても `None` を返すだけで例外を投げない。必ず `None` チェックを行う。

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import DATA_DIR, load_image, show_nb

IMG_PATH = DATA_DIR / 'lena.png'
print(f'画像パス: {IMG_PATH}')
print(f'存在: {IMG_PATH.exists()}')
```

```python
# --- 読み込み ---
color = load_image(IMG_PATH, cv2.IMREAD_COLOR)       # BGR 3ch
gray  = load_image(IMG_PATH, cv2.IMREAD_GRAYSCALE)   # 1ch
bgra  = load_image(IMG_PATH, cv2.IMREAD_UNCHANGED)   # アルファがあれば 4ch

print(f'color shape : {color.shape}  dtype: {color.dtype}')
print(f'gray  shape : {gray.shape}   dtype: {gray.dtype}')
print(f'bgra  shape : {bgra.shape}   dtype: {bgra.dtype}')
```

```python
show_nb([
    ('color (BGR→RGB)', color),
    ('gray', gray),
    ('bgra/unchanged', bgra),
], cols=3)
```

```python
# --- 保存 ---
out_path = Path('01_io_out.png')
cv2.imwrite(str(out_path), color)
print(f'saved: {out_path.resolve()}')

# 読み直して確認
reloaded = cv2.imread(str(out_path))
print(f'reloaded shape: {reloaded.shape}')
```

```python
# --- ピクセル値の操作 ---
px = color[100, 100]   # BGR
print(f'pixel (100,100) BGR = {px}')

# NumPy スライスで ROI 取得
roi = color[50:150, 50:150]
show_nb([('ROI 100x100', roi)], cols=1)
```
