# 07 — モルフォロジー演算

## 目標
- 基本演算: 収縮 (erosion) / 膨張 (dilation)
- 派生演算: Opening / Closing / Gradient / Top Hat / Black Hat

## 演算まとめ
| 演算 | 定義 | 効果 |
|------|------|------|
| Erosion | min フィルタ | 前景を縮小、ノイズ除去 |
| Dilation | max フィルタ | 前景を拡大、穴埋め |
| Opening | erosion→dilation | 細突起除去 |
| Closing | dilation→erosion | 小さな穴埋め |
| Gradient | dilation-erosion | 輪郭抽出 |
| Top Hat | src-opening | 明るい小物体 |
| Black Hat | closing-src | 暗い小物体 |

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
import numpy as np
from utils import DATA_DIR, load_image, show_nb

src  = load_image(DATA_DIR / 'lena.png', cv2.IMREAD_GRAYSCALE)
_, binary = cv2.threshold(src, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

k3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
k7 = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
ke = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (9, 9))
```

```python
eroded  = cv2.erode(binary, k3, iterations=2)
dilated = cv2.dilate(binary, k3, iterations=2)

show_nb([
    ('binary (Otsu)', binary),
    ('erode ×2', eroded),
    ('dilate ×2', dilated),
], cols=3)
```

```python
opened   = cv2.morphologyEx(binary, cv2.MORPH_OPEN,     k7)
closed   = cv2.morphologyEx(binary, cv2.MORPH_CLOSE,    k7)
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, k3)
tophat   = cv2.morphologyEx(src,    cv2.MORPH_TOPHAT,   ke)
blackhat = cv2.morphologyEx(src,    cv2.MORPH_BLACKHAT, ke)

show_nb([
    ('opening 7×7', opened),
    ('closing 7×7', closed),
    ('gradient', gradient),
    ('top hat', tophat),
    ('black hat', blackhat),
], cols=3)
```
