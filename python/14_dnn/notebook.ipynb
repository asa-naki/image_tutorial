モデルがない場合はエッジ検出デモが表示されます。

````python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
import numpy as np
from utils import DATA_DIR, load_image, show_nb
from main import DEFAULT_MODEL, run

IMG_PATH = DATA_DIR / 'lena.png'
print(f'モデルパス: {DEFAULT_MODEL}')
print(f'モデル存在: {DEFAULT_MODEL.exists()}')
````

````python
pairs = run(IMG_PATH, conf=0.4)
show_nb(pairs, cols=2)
````

````python
# blobFromImage の前処理を手動で確認
src = load_image(IMG_PATH)
blob = cv2.dnn.blobFromImage(src, 1/255.0, (640, 640), swapRB=True, crop=False)
print(f'blob shape: {blob.shape}')   # (1, 3, 640, 640)
print(f'blob range: [{blob.min():.3f}, {blob.max():.3f}]')

# blob の最初のチャネルを可視化
ch0 = (blob[0, 0] * 255).astype(np.uint8)
show_nb([('blob channel 0 (R)', ch0)], cols=1)
````
