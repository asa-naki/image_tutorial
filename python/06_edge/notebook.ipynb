# 06 — エッジ検出

## 目標
- Sobel / Scharr で勾配を求める
- Laplacian で2次微分エッジを検出
- Canny で高精度エッジ検出 (ヒステリシス + NMS)

## アルゴリズム概要
- **Sobel**: X/Y方向の1次微分カーネル。大きさ `sqrt(Gx²+Gy²)`
- **Canny**: ①ガウシアン ②Sobel勾配 ③Non-Maximum Suppression ④ヒステリシス閾値

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
import numpy as np
from utils import DATA_DIR, load_image, show_nb

src  = load_image(DATA_DIR / 'lena.png')
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
# Sobel
Gx = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=3)
Gy = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=3)
mag = np.clip(cv2.magnitude(Gx, Gy), 0, 255).astype(np.uint8)

# Scharr (Sobelより精度高)
Sx = cv2.Scharr(blur, cv2.CV_64F, 1, 0)
Sy = cv2.Scharr(blur, cv2.CV_64F, 0, 1)
scharr_mag = np.clip(cv2.magnitude(Sx, Sy), 0, 255).astype(np.uint8)

show_nb([
    ('gray', gray),
    ('Sobel X', cv2.convertScaleAbs(Gx)),
    ('Sobel Y', cv2.convertScaleAbs(Gy)),
    ('Sobel magnitude', mag),
    ('Scharr magnitude', scharr_mag),
], cols=3)
```

```python
# Laplacian
lap = cv2.convertScaleAbs(cv2.Laplacian(blur, cv2.CV_64F, ksize=3))

# Canny (閾値を変えて比較)
canny1 = cv2.Canny(blur,  50, 150)
canny2 = cv2.Canny(blur, 100, 200)

show_nb([
    ('Laplacian', lap),
    ('Canny (50, 150)', canny1),
    ('Canny (100, 200)', canny2),
], cols=3)
```
