# 02 — 色空間変換

## 目標
- BGR ↔ Gray / HSV / Lab の変換
- `cv2.split` でチャネル分離
- HSV で色相範囲を指定してマスクを作る

## 色空間まとめ
| 空間 | 用途 |
|------|------|
| BGR  | OpenCV デフォルト |
| Gray | 輝度のみ。計算コスト削減 |
| HSV  | 色相 (H)・彩度 (S)・明度 (V) で直感的な色選択が可能 |
| Lab  | 知覚均一。色差計算 (ΔE) に使う |


```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import DATA_DIR, load_image, show_nb

IMG_PATH = DATA_DIR / 'lena.png'
bgr = load_image(IMG_PATH)
```

```python
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
hsv  = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
lab  = cv2.cvtColor(bgr, cv2.COLOR_BGR2Lab)
b, g, r = cv2.split(bgr)

show_nb([
    ('BGR (→RGB表示)', bgr),
    ('Gray', gray),
    ('HSV', hsv),
    ('Lab', lab),
    ('B channel', b),
    ('G channel', g),
    ('R channel', r),
], cols=4)
```

```python
# HSV で赤色を抽出 (H=0付近 と H=180付近)
mask1 = cv2.inRange(hsv, (0,  80, 80), (10,  255, 255))
mask2 = cv2.inRange(hsv, (170, 80, 80), (180, 255, 255))
red_mask = mask1 | mask2
red_only = cv2.bitwise_and(bgr, bgr, mask=red_mask)

show_nb([
    ('red mask', red_mask),
    ('red only', red_only),
], cols=2)
```

```python
# チャネルを合成して BGR に戻す
merged = cv2.merge([b, g, r])
print(f'元の画像と差分 max: {abs(bgr.astype(int) - merged.astype(int)).max()}')
show_nb([('merged (=original)', merged)], cols=1)
```
