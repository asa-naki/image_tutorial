# 04 — 閾値処理・二値化

## 目標
- 固定閾値 / Otsu / Triangle による大域二値化
- `adaptiveThreshold` による局所的二値化

## 使い分け
| 手法 | 向いている状況 |
|------|---------------|
| 固定閾値 | 照明が均一で閾値が既知 |
| Otsu | グレーヒストグラムが双峰性 |
| Adaptive | 照明むらがある文書など |


```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import cv2
from utils import DATA_DIR, load_image, show_nb

gray = load_image(DATA_DIR / 'lena.png', cv2.IMREAD_GRAYSCALE)
```

```python
# 固定閾値
_, fixed = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Otsu
otsu_thr, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print(f'Otsu 閾値: {otsu_thr:.1f}')

# Triangle
tri_thr, triangle = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_TRIANGLE)
print(f'Triangle 閾値: {tri_thr:.1f}')

show_nb([
    ('gray', gray),
    ('fixed thr=127', fixed),
    (f'Otsu thr={otsu_thr:.0f}', otsu),
    (f'Triangle thr={tri_thr:.0f}', triangle),
], cols=2)
```

```python
# Adaptive Threshold
adapt_mean  = cv2.adaptiveThreshold(gray, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 5)
adapt_gauss = cv2.adaptiveThreshold(gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 5)

show_nb([
    ('gray', gray),
    ('adaptive mean C=5', adapt_mean),
    ('adaptive Gaussian C=5', adapt_gauss),
], cols=3)
```
